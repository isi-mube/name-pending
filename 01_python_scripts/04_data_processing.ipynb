<h2 style="color: #008080;">Dealing with Data distributions</h2>

```python
# set color palette
colors = ['#0072B2'] # blue

# plot normalized features in a grid
fig, axes = plt.subplots(nrows=5, ncols=5, figsize=(20, 16)) # nrows to 5, ncols to 6 to plot our 26 features
for i, ax in enumerate(axes.flat):
    if i >= len(cleaned.columns):
        break
    sns.histplot(cleaned.iloc[:, i], ax=ax, color=colors[i % len(colors)], kde=True)
    sns.kdeplot(cleaned.iloc[:, i], ax=ax, color='#999999', linewidth=2)
    ax.set_title(cleaned.columns[i])
plt.tight_layout()
plt.show()
```
